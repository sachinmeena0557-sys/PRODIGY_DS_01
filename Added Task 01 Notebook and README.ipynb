```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV file load karo
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# Sirf Country Name aur 2023 column lo
df = df[['Country Name', '2023']]

# Empty rows hatao
df = df.dropna()

# Top 10 countries select karo
top10 = df.sort_values(by='2023', ascending=False).head(10)

# Graph banao
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2023'])

plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Population (2023)")
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# Aggregate rows remove
exclude = ['World','IDA & IBRD total','Low & middle income',
           'Middle income','IBRD only','Early-demographic dividend',
           'Late-demographic dividend','East Asia & Pacific',
           'South Asia']

df = df[~df['Country Name'].isin(exclude)]

# Select columns
df = df[['Country Name','2023']].dropna()

# Top 10 countries
top10 = df.sort_values(by='2023', ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2023'])

plt.xlabel("Countries")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2023)")
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset load
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# Sirf country rows manually select
countries = ['India','China','United States','Indonesia','Pakistan',
             'Nigeria','Brazil','Bangladesh','Russia','Mexico']

df = df[df['Country Name'].isin(countries)]

# Select columns
df = df[['Country Name','2023']].dropna()

# Sort
df = df.sort_values(by='2023', ascending=False)

# Plot
plt.figure(figsize=(12,6))
plt.bar(df['Country Name'], df['2023'])

plt.xlabel("Countries")
plt.ylabel("Population")
plt.title("Population Distribution of Top 10 Countries (2023)")
plt.xticks(rotation=45)

plt.show()
```
